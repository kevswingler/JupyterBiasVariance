```python
%matplotlib notebook
from pylab import *
import numpy as np
import matplotlib
import math

from matplotlib import pyplot as plt


x = np.arange(0, 100)
y = np.sin(x/30) + np.random.normal(0, 0.2, 100)
plt.plot(x, y, "*")
```

```python
xs=np.random.choice(x,10,False)
ys=y[xs]
plt.plot(xs, ys, "o")
```

```python
linear=polyfit(xs,ys,1)
linear_fun=np.poly1d(linear)
plt.plot(x,linear_fun(x))
```

```python
order2=polyfit(xs,ys,2)
order2_fun=np.poly1d(order2)
plt.plot(x,order2_fun(x))
```

```python
order9=polyfit(xs,ys,9)
order9_fun=np.poly1d(order9)
plt.plot(x,order9_fun(x))
axes=plt.gca()
axes.set_ylim([-2,2])
```
